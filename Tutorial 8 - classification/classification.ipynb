```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("insurance_data.csv")
df = pd.DataFrame(data)

sns.pairplot(df)
```

```python
# For classification problems, we either use sigmoid function
# sigmoid(z) = 1 / 1 + e ^ (-z) where z = mx+b
# and LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['bought_insurance'], axis='columns'), df.drop(['age'], axis='columns'), test_size=0.1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
print(model.score(X_test,y_test))
X_test
```

```python
model.predict_proba(X_test)
```
