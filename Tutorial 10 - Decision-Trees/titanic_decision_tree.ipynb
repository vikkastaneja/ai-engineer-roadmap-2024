```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("titanic.csv")
df
```

```python
# First convert the non-numeric field to numeric
dummies = pd.get_dummies(df['Sex'])
dummies
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
df['sex_n'] = le_sex.fit_transform(df['Sex'])
df
```

```python
df_new = df.drop(['Sex'], axis='columns')
df_new
```

```python

X = df_new[['Pclass', 'Age', 'Fare', 'sex_n']]
X


X['Age'] = X['Age'].fillna(X['Age'].mean())
X.isna().sum()
```

```python
X.head()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, df_new['Survived'])
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True)
plt.xlabel("Precited")
plt.ylabel("Truth")
```
