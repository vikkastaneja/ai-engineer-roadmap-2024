```python
import numpy as np
import pandas as pd

data = pd.read_csv("carprices.csv")
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.DataFrame(data)
df
# plt.scatter(df['Age(yrs)'], df['Mileage'], marker='x', color='red')

# plt.scatter(df['Age(yrs)'], df['Sell Price($)'], marker='x', color='blue')
# plt.scatter(df['Mileage'], df['Sell Price($)'], marker='o', color='green')
# plt.xlabel('Age of Car', fontsize=15)
# plt.ylabel("Mileage of Car", fontsize=15)
sns.pairplot(df)
plt.show()
```

```python
# Since the plot is close to linear, price can be predicted from age and/or Mileage
dummies = pd.get_dummies(df['Car Model'], dtype=int)
dummies
```

```python
# Concat the columns of dummies with actual
df_dummies = pd.concat([df, dummies], axis=1)
df_dummies
```

```python
# Drop text column and one car to avoid dummy variable trap
df_dummies.drop(['Car Model', 'Mercedez Benz C class'], axis='columns', inplace=True)
df_dummies
```

```python
X = df_dummies.drop(['Sell Price($)'], axis='columns')
y = df_dummies['Sell Price($)']
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y) 
```

```python
model.predict([[45000, 4, 0, 0]])
```

```python
model.predict([[86000, 7, 0, 1]])
```

```python
model.score(X, y)
```
